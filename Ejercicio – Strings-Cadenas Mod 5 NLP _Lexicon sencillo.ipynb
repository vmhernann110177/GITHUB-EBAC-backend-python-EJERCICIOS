```python
## Analizador de sentimiento -----

### 1. EJEMPLO PARA SENTIMIENTO POSITIVO:

# Cadena a tratar, OJO modifiqué ligeramente la cadena:
# pln_string = "Este es un producto muy bueno, excelente, maravilloso, pero incompleto y básico"
nlp_string = """Este es un producto muy bueno, excelente, maravilloso, pero incompleto y básico"""

# 1. Quiero ver, en terminos de caracteres, algunas caracteristicas de la cadena que se lee:

print("\nQuiero ver, en terminos de caracteres, \nalgunas caracteristicas de la cadena que se lee:\n")
print("Todos los caracteres de la cadena son = \n\"", nlp_string[:],"\"")
print("\nEl número de caracteres que leo es =", len(nlp_string)) # esta leyendo 80 caracteres
print("El Caracter inicial es = |", nlp_string[0],"|")
print("El Caracter final es = |", nlp_string[-1],"|")

# CÓDIGO MODIFICADO:

my_string = nlp_string

neutral_words = ["producto", "básico"]
positive_words = ["bueno", "excelente", "maravilloso"]
negative_words = ["defectuoso", "malo", "horrible", "incompleto"]

positive_score = 0
negative_score = 0
neutral_score = 0
Result =""

for positive_word in positive_words:
    positive_score = positive_score + my_string.count(positive_word)

for negative_word in negative_words:
    negative_score = negative_score - my_string.count(negative_word)

for neutral_word in neutral_words:
    neutral_score = neutral_score + my_string.count(neutral_word)

print("\n\u001b[34mPRESENTACIÓN:" + "\033[0m")    
print( "Esto es un \033[1m" + "Analizador \u001b[31m MUY SENCILLO \u001b[30m de sentimiento \u001b[34m(Basado en un corto lexicón)." + "\033[0m")    
print("\nmy_string: \"{}\" ".format(my_string))
print("\nscore positivo, resultado:{}, en valor absoluto:{}".format(positive_score, abs(positive_score)))
print("\nscore negativo, resultado:{}, en valor absoluto:{}".format(positive_score, abs(negative_score)))
print("\nscore Neutro, resultado:{}, en valor absoluto:{}".format(positive_score, abs(neutral_score)))

if (abs(positive_score)>abs(negative_score) and abs(positive_score)>abs(neutral_score)):
    Result = "Positivo '+'"
    
if (abs(negative_score)>abs(positive_score) and abs(negative_score)>abs(neutral_score)):
    Result = "Negativo '-'"
    
if (abs(neutral_score)>abs(positive_score) and abs(neutral_score)>abs(negative_score)):
    Result = "Neutro '0'"
    
print("\nEn general se puede decir que el sentimiento de esta oracion es: {}".format(Result),"\n")
```
